# Calculate the speed at a given chainring-sprocket ratio

While using regular road bikes with a shifting system, one generally rides in a ratio that feels most comfortable, and is interestingly most of the times the most efficient. 
However, if you are riding a fixed gear, cadence and speed are inseparably connected. To know which ratio is optimal, one should strike for a cadence of 80-100, which can be quite fast depending on the chainring and ratio.

This Notebook aims to give more details into speed-cadence.

```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
```

```python
durchmesser = 28 * 2.54 / 100

Umfang = durchmesser * np.pi

def geschwindigkeit(n_ritzel, n_kettenblatt):
  ratio = n_kettenblatt/n_ritzel
  # zurückgelegte Strecke pro Kurbelumdrehung
  strecke = ratio * Umfang
  # Kadenz als Umdrehungen pro Minute
  cadence = np.arange(50,141) 
  # zurückgelegte Strecke pro Minute
  strecke_pro_stunde = cadence * strecke
  # Umrechnung in km/h
  km_pro_stunde = strecke_pro_stunde * 60 / 1000
  return km_pro_stunde

km_h = geschwindigkeit(16, 49)
cadence = np.arange(50,141) 

plt.figure(figsize=(10,5))
plt.plot(cadence, km_h)
plt.xlabel("Kadenz (U/min)")
plt.ylabel("Geschwindigkeit (km/h)")
plt.title("Geschwindigkeit in Abhängigkeit von der Kadenz")
plt.legend
plt.grid(True)
plt.show()

  

```

```python
n_kettenblatt = 49
n_ritzel_list = [14, 15, 16, 17, 18]

# Plot vorbereiten
plt.figure(figsize=(10, 5))

for n_ritzel in n_ritzel_list:
    geschw = geschwindigkeit(n_ritzel, n_kettenblatt)
    ratio = round(n_kettenblatt / n_ritzel, 2)
    plt.plot(cadence, geschw, label=f"{n_kettenblatt}/{n_ritzel} ({ratio})")

# Plot-Details
plt.xlabel("Kadenz (U/min)")
plt.ylabel("Geschwindigkeit (km/h)")
plt.title("Geschwindigkeit bei verschiedenen Übersetzungen")
plt.legend(title="Übersetzung\n(Kettenblatt/Ritzel)")
plt.grid(True)
plt.show()
```

```python
def plot_interaktiv(n_ritzel=15, n_kettenblatt=49):
    geschw = geschwindigkeit(n_ritzel, n_kettenblatt)
    kadenzen = np.arange(50, 141)

    plt.figure(figsize=(11, 7))
    plt.plot(kadenzen, geschw, color='blue')
    plt.xlabel("Kadenz (U/min)")
    plt.ylabel("Geschwindigkeit (km/h)")
    plt.title(f"n_Ritzel = {n_ritzel}, n_Kettenblatt = {n_kettenblatt}")
    plt.grid(True)
    plt.ylim(0, 60)
    plt.show()

interact(plot_interaktiv, 
         n_ritzel=IntSlider(min=10, max=25, step=1, value=15, description='Ritzel'),
         n_kettenblatt=IntSlider(min=30, max=60, step=1, value=49, description='Kettenblatt'));
```
